**THE PULSE OF THE CITY **

ACT-1: A Number That Chooses the Story  

```python
import sympy as sp

# 1) Enter last 4 digits
W = 1
X = 2
Y = 3
Z = 4

# 2) Digital root
def digital_root(n):
    while n > 9:
        n = sum(int(d) for d in str(n))
    return n

R = digital_root(W + X + Y + Z)

# 3) Define variable
x = sp.symbols('x')

# 4) Define function f(x)
f = (x**3)/3 - R*(x**2) + (R**2 - 1)*x

# 5) Differentiate
f_prime = sp.diff(f, x)

# 6) Solve f'(x) = 0
critical_points = sp.solve(f_prime, x)

# 7) Output
print("R =", R)
print("f'(x) =", f_prime)
print("Critical Points:", critical_points)
```

“The critical points at 0 and 2 represent transition zones where system behavior changes. These define the focused range for further analysis instead of considering the entire dataset.”

ACT-2: The Raw City

```python
import pandas as pd

df = pd.read_csv("yellow_tripdata_2015-01.csv")  # exact name
df.head()
```

“The dataset contains detailed trip-level information including timestamps, distance, fare, and location. Initial inspection shows that while the structure appears complete, there may exist inconsistencies such as very short trips, zero tip amounts, and potential anomalies in fare and distance values. Further analysis is required to identify invalid or unrealistic trips.”

```python
df.shape
df.columns
```

```python
df.describe()
```

“The dataset contains invalid values such as zero or negative distances, fares, and passenger counts, along with extreme outliers. High variability and inconsistent values indicate the presence of erroneous records, making data cleaning necessary.”

```python
df.isnull().sum()
```

“No significant missing values were observed in the dataset

```python
df_invalid = df[
    (df['trip_distance'] <= 0) |
    (df['fare_amount'] <= 0) |
    (df['passenger_count'] <= 0)
]

df_invalid.shape
```

“A total of 9013 records were identified as invalid based on conditions such as non-positive trip distance, fare amount, or passenger count. This confirms the presence of unrealistic trips within the dataset that do not represent actual taxi activity.”

```python
df_clean = df[
    (df['trip_distance'] > 0) &
    (df['fare_amount'] > 0) &
    (df['passenger_count'] > 0) &
    (df['tpep_dropoff_datetime'] > df['tpep_pickup_datetime'])
]
```

```python
print("Original:", len(df))
print("Cleaned:", len(df_clean))
```

“Although the number of removed records is relatively small compared to the total dataset, their presence can still impact analysis, especially when identifying patterns or anomalies.                                      “Although the dataset contains no missing values, the presence of invalid and unrealistic records highlights that completeness does not guarantee data quality. Careful validation is therefore essential before analysis.”

“Even a small proportion of invalid records can significantly affect aggregated metrics such as average fare or trip distance, potentially leading to misleading conclusions.”

**ACT-5: Pressure on the System**

“Fare amount is selected as the key parameter because it directly influences passenger behavior, revenue, and tipping patterns, making it a critical factor in the system.”

```python
#Applying +1% change as the R value considered in ACT-1 is = 1
R = 1

df_sim = df_clean.copy()

# Increase fare by R%
df_sim['fare_amount_new'] = df_sim['fare_amount'] * (1 + R/100)

df_sim[['fare_amount', 'fare_amount_new']].head()
```

“A 1% increase was applied to the fare amount, resulting in a proportional increase in all fare values. The transformation is consistent across all records, confirming that the simulation has been correctly implemented.”

```python
df_clean['fare_amount'].mean(), df_sim['fare_amount_new'].mean()
```

```python
df_clean['fare_amount'].sum(), df_sim['fare_amount_new'].sum()
```

“The average fare increased slightly after applying the 1% change, indicating a proportional adjustment across all trips. The small magnitude of change suggests that the overall pricing structure of the system remains stable and is not significantly disrupted.”


“The total revenue shows a noticeable increase following the fare adjustment, confirming that even a small percentage change can lead to a significant cumulative impact at scale. This highlights the sensitivity of total system revenue to pricing changes.”


“The system absorbs the change in terms of behavior, but amplifies it in terms of aggregate revenue.”
